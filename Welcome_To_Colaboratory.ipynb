# Non-Linear

```
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# def tanh(x):
#     return np.tanh(x)

def tanh(x):
  return ((np.exp(x))-(np.exp(-x)))/((np.exp(-x))+(np.exp(x)))

def relu(x):
    return np.maximum(0, x)

def leakyrelu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / np.sum(exp_x, axis=0)

x = np.linspace(-5, 5, 100)

sigmoid_values = sigmoid(x)
tanh_values = tanh(x)
relu_values = relu(x)
leaky_relu_values = leakyrelu(x)
softmax_values = softmax(x)

plt.figure(figsize=(10, 6),facecolor='lightgray')

plt.subplot(3, 2, 1)

plt.plot(x, sigmoid_values, label='Sigmoid',color ='red')
plt.legend()
plt.title('Sigmoid')

plt.subplot(3, 2, 2)
plt.plot(x, tanh_values, label='Tanh',color ='red')
plt.legend()
plt.title('Tanh')

plt.subplot(3, 2, 3)
plt.plot(x, relu_values, label='ReLU',color ='red')
plt.legend()
plt.title('ReLU')

plt.subplot(3, 2, 4)
plt.plot(x, leaky_relu_values, label='Leaky ReLU',color ='red')
plt.legend()
plt.title('Leaky ReLU')

plt.subplot(3,2, 5)
plt.plot(x,softmax_values , label='Softmax',color ='red')
plt.legend()
plt.title('Softmax')

plt.tight_layout()
plt.show()
```

# Linear

```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)

linear_value = x
plt.figure(figsize=(6, 6),facecolor='lightgray')

ax=plt.axes()
ax.set_facecolor("lightgrey")

plt.plot(x, linear_value , label='Linear',color= 'blue')
plt.legend()
plt.title('Linear')

plt.tight_layout()
plt.show()

```

# Binary

```
import numpy as np
import matplotlib.pyplot as plt

def bin(x):
     return np.where(x>=0,1,0)

x = np.linspace(-5, 5, 100)

y = bin(x)


plt.figure(figsize=(6, 6),facecolor='lightgray')

ax=plt.axes()
ax.set_facecolor("lightgrey")

plt.plot(x, y , label='Binary',color='blue')
plt.legend()
plt.title('Binary')


plt.tight_layout()
plt.show()

```
